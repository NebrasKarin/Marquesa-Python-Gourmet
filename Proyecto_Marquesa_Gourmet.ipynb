# ü•ß Marquesa de Parchita Gourmet (Receta Probada)
Esta receta es equilibrada (no empalaga) y tiene la consistencia perfecta sin usar batidora.

### üõí Ingredientes:
* **L√°cteos:** 170g de crema de leche (**Di Fiore**) y 170g de leche condensada (**Natulac**).
* **Fruta:** 500g de pulpa de parchita con semillas.
* **Galletas:** 32 galletas Mar√≠a (**Puig**) (8 paquetes individuales).

### üë©‚Äçüç≥ Preparaci√≥n Paso a Paso:
1. **Parchita:** Licuar 350g por 2 segundos y colar. Reservar 150g con semillas para decorar.
2. **Mousse:** Licuar l√°cteos fr√≠os y verter el jugo en hilo hasta espesar.
3. **Armado:** Capas de crema y galletas (aprox. 7 por capa). Salen 4 o 5 capas.
4. **Decoraci√≥n:** Cocinar la pulpa reservada con az√∫car y verter fr√≠a.
5. **Reposo:** M√≠nimo 6 horas en nevera.

```python
# --- C√ÅLCULOS T√âCNICOS DE LA MARQUESA ---
ingredientes = {
    "crema_leche": 170,
    "leche_condensada": 170,
    "pulpa_total": 500,
    "galletas_totales": 32
}

# Separaci√≥n de la fruta seg√∫n tu receta
jugo_mousse = 350
decoracion = ingredientes["pulpa_total"] - jugo_mousse

# C√°lculo de capas
galletas_por_capa = 7
num_capas = ingredientes["galletas_totales"] // galletas_por_capa

print(f"FICHA T√âCNICA GENERADA:")
print(f"------------------------")
print(f"Separar {jugo_mousse}g de jugo para la mezcla.")
print(f"Reservar {decoracion}g de pulpa para el alm√≠bar final.")
print(f"Con {ingredientes['galletas_totales']} galletas har√°s {num_capas} capas completas.")
```
